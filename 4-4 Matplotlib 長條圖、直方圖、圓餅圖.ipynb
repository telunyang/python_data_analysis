# 長條圖 bar()
- 通常用於「類別型」資料 (categorical data)
  - 資料之間是離散的 (discrete)、間斷的
- plt.bar(x, height, width=0.8, ...)
  - `x`: array-like 的資料，例如 list、array、tuple 等，作為 x 軸的標題
  - `height`: array-like 的資料，作為 y 軸的值
  - `width`: bar 的寬度
- matplotlib.pyplot.xticks(ticks=None, labels=None, ...)
  - `ticks`: x 軸的刻度
  - `label`: 對應 x 軸的標題，類似取代 x 軸的刻度，用其它名稱代替
- 參考連結:
  - [matplotlib.pyplot.bar](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html)
  - [matplotlib.pyplot.xticks](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html)
  - [Python matplotlib 畫圖 – 柱狀圖 (長條圖)](https://python-ecw.com/2020/11/12/matplotlib柱狀圖/)

```python
'''
班上投票，Alex 得票 7749，Bill 得票 9527，Carl 得票 2266，請用長條圖表示
'''

# x 軸文字的座標
x = [1, 2, 3]

# 得票數 (因為有 3 名候選人)
height = [7749, 9527, 2266]
candidates = ['Alex', 'Bill', 'Carl']

# bar 的寬度
width = 0.5

# 繪製長條圖
plt.bar(
    x, 
    height, 
    width, 
    edgecolor='black', 
    linewidth=1, 
    color=['yellow', 'green', 'red']
)

# 設定數值文字在圖表上
for index in range(len(x)):
    # matplotlib.pyplot.text(圖表的x座標, 圖表的y座標, 顯示的文字)
    plt.text(
        x[index], 
        height[index], 
        height[index])

# 設定圖表標題
plt.title('Election')
plt.ylabel('Votes')

# 設定 x 和 y 軸的刻度
plt.xticks(x, candidates)
plt.yticks(np.arange(0, 10000, step=1000))

# 顯示圖表
plt.show()
```

# 直方圖
- 通常用於「數值型」資料 (numeric data)
  - 資料之間是連續的 (continueous)
- plt.hist(x, bins, ...)
  - x: 你的資料，list、array 或是 sequece
  - bin: 組距，預設 10
- 參考資料:
  - [matplotlib.pyplot.hist](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)
  - [Python 學習筆記 : Matplotlib 資料視覺化 (二) 統計圖](http://yhhuang1966.blogspot.com/2020/05/python-matplotlib.html)

```python
'''
請以班上 100 個學生考試成績，最低 0 分，最高 100 分，繪製直方圖
'''

# 隨機產生 100 位同學的成績 (用 np.random.randint 的話，要轉成 list)
x = list(np.random.randint(0, 101, size=100))

# 組距 (將成績 0 到 100 分，組距為 10，回傳 11 個數字)
bins = np.arange(0, 101, 10)

# 繪製直方圖 (會回傳每個組距的數量，以及對應組距)
arr_counts, arr_bins, obj = plt.hist(
    x, 
    bins,
    edgecolor='black',
    linewidth=1
)

# 設定數值文字在圖表上
for i in range(len(arr_counts)):
    plt.text(
        arr_bins[i], # x 為 0 ~ 90 的座標上寫字
        arr_counts[i], 
        arr_counts[i].astype('int32')
    )

# 設定標題
plt.title('Students Exam')
plt.xticks(bins)
plt.ylabel('Scores')

# 顯示圖表
plt.show()
```

# 圓餅圖 pie()
- 用於描述「類別型」資料
- plt.pie(x, explode=None, labels=None, colors=None, autopct=None,shadow=False, labeldistance=1.1, radius=1, counterclock=True, center=(0, 0), ...)
  - x: 你的資料。
  - explode: 餅分離的程度，list 格式，其中的每個元素值介於 0 到 1 之間。
  - labels: 圓餅圖的標題，list 格式。
  - autopct: 每一個項目的百分比格式，例如「%2.2f%%」，代表整數 2 位數，小數 2 位數。
  - labeldistance: 項目標題與圓餅圖中間距離是半徑的多少倍。
  - center: 圓中心座標，預設 (0, 0)。
  - shadow: 圓餅圖是否有陰影，預設為 False。
- 參考資料:
  - [matplotlib.pyplot.pie](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html#matplotlib.pyplot.pie)

```python
'''
有一家庭支出費用如下，以此設計圓餅圖:
旅行 (Travel): 8000
娛樂 (Entertainment): 2000
教育 (Education): 3000
交通 (Transportation): 5000
餐費 (Food): 6000
'''

# 設定支出項目
labels = [
    'Travel', 
    'Entertainment', 
    'Education', 
    'Transportation', 
    'Food'
]

# 設定支出費用
x = [
    8000, 
    2000, 
    3000, 
    5000, 
    6000
]

# 繪製圓餅圖
plt.pie(
    x,
    explode=(0, 0.2, 0, 0, 0),
    labels=labels,
    autopct="%.2f%%"
)

# 顯示圖表
plt.show()
```

# 其它範例

```python
'''Stacked Bar chart'''

# 匯入套件
import matplotlib.pyplot as plt
import numpy as np

# 設定字型 (optional)
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')

# 建立資料
'''
y 用 np.array() 來產生數據，
是因為 plt.bar() 當中的 bottom 參數，
可以透過與上一個陣列相加來賦值
'''
x = ['A', 'B', 'C', 'D']
h1 = np.array([10, 20, 10, 30])
h2 = np.array([20, 25, 15, 25])
h3 = np.array([12, 15, 19, 6])
h4 = np.array([10, 29, 13, 19])
 
# 繪製 stacked bar
# (重點在 bottom 屬性; h1 和 h2 h3 可以相加，是因為 numpy)
plt.bar(x, h1, color='r', edgecolor='black', linewidth=1)
plt.bar(x, h2, bottom=h1, color='b', edgecolor='black', linewidth=1)
plt.bar(x, h3, bottom=h1+h2, color='y', edgecolor='black', linewidth=1)
plt.bar(x, h4, bottom=h1+h2+h3, color='g', edgecolor='black', linewidth=1)

# 圖表標題
plt.xlabel("隊伍")
plt.ylabel("分數")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("每個隊伍在 4 輪中的分數")

# 顯示圖表
plt.show()
```

```python
'''
Horizontal Bar chart
'''
# 匯入套件
import matplotlib.pyplot as plt

# 設定字型 (optional)
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')

# 設定類別項目
list_labels = [
    'Python', 
    'Scala', 
    'C#', 
    'Java', 
    'PHP'
]

# 設定類別項目的 x 軸座標
y = [1, 2, 3, 4, 5]

# 每個類別項目的值
width = [50, 10, 13, 27, 22]

# 繪製 Horizontal Bar (y 標籤順序由下往上)
plt.barh(
    y, 
    width,
    edgecolor='black',
    linewidth=1, 
    tick_label=list_labels, 
    color=['red', 'green', 'blue', 'yellow', 'purple'])

# 設定數值文字在圖表上
for i in range(len(x)):
    # plt.text() 依然使用 x, y text 的順序
    plt.text(width[i], y[i], width[i])

# 圖表標題
plt.ylabel("程式語言")
plt.xlabel("使用次數")

# 顯示圖表
plt.show()
```

```python
'''
Grouped Bar chart
'''
# 匯入套件
import matplotlib.pyplot as plt

# 設定字型 (optional)
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')

# 設定群組資料的值 (第1隊、第2隊)
y_score_1 = [35, 60, 18, 29, 77]
y_score_2 = [28, 33, 93, 44, 32]

# 設定 x 軸的順序，這裡要注意的是，group 的 bar chart，要另外指定 x 軸的位置
BAR_WIDTH = 0.4
x_axis_1 = [1, 2, 3, 4, 5]
x_axis_2 = [i + BAR_WIDTH for i in x_axis_1]
'''
x_axis_2 = [
    1 + BAR_WIDTH,
    2 + BAR_WIDTH,
    3 + BAR_WIDTH,
    4 + BAR_WIDTH,
    5 + BAR_WIDTH
]
'''

# 設定 x 刻度的中心位置
ticks = [i + BAR_WIDTH / 2 for i in x_axis_1]
'''
ticks = [
    1 + BAR_WIDTH / 2,
    2 + BAR_WIDTH / 2,
    3 + BAR_WIDTH / 2,
    4 + BAR_WIDTH / 2,
    5 + BAR_WIDTH / 2
]
'''

# 設定 x 標題
list_labels = ["Round 1", "Round 2", "Round 3", "Round 4", "Round 5"]

# 繪製圖表
plt.bar(
    x_axis_1, 
    y_score_1, 
    BAR_WIDTH, 
    color="y", 
    edgecolor='black',
    linewidth=1,
    label="第 1 隊"
)
plt.bar(
    x_axis_2, 
    y_score_2, 
    BAR_WIDTH, 
    color="b", 
    edgecolor='black',
    linewidth=1,
    label="第 2 隊"
)

# 設定數值文字在圖表上
for i in range(len(x_axis_1)):
    plt.text(
        x_axis_1[i], 
        y_score_1[i], 
        y_score_1[i]
    )
    plt.text(
        x_axis_2[i], 
        y_score_2[i], 
        y_score_2[i]
    )

# 圖表標題
plt.xlabel("回合數")
plt.ylabel("分數")
plt.title("隊伍競賽分數")
plt.xticks(ticks, list_labels)
plt.legend(loc='upper left', ncols=2)

# 顯示圖表
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 匯入資料。記得先安裝 openpyxl
df = pd.read_excel('./src01.xlsx')

# 資料
x_name = df['name']
y_num01 = df['num01']
y_num02 = df['num02']
y_num03 = df['num03']
y_num04 = df['num04']

# 設定每組的間距
bar_width = 0.2  
x_indices = np.arange(len(x_name))

# 建立兩個 subplot，共享 x 軸
plt.figure(figsize=(10, 10), dpi=100)

# 上方長條圖 (num01, num02, num03)
plt.subplot(2, 1, 1)
plt.bar(x_indices - 0.2, y_num01, width=bar_width, label='num01')
plt.bar(x_indices, y_num02, width=bar_width, label='num02')
plt.bar(x_indices + 0.2, y_num03, width=bar_width, label='num03')
plt.xticks([])  # 移除 x 軸標籤 (保留下方)
plt.ylabel('y_label_for_01_02_03')  # y 軸標籤 (上方)
plt.title('Title01')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# 下方長條圖 (num04) - 值較小獨立展示
plt.subplot(2, 1, 2)
plt.bar(x_indices, y_num04, width=bar_width, color='orange', label='num04')
plt.xticks(x_indices, x_name)
plt.xlabel('Name')
plt.ylabel('y_label_for_04')  # y 軸標籤 (下方)
plt.title('Title02')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# 調整子圖之間的間距。tight_layout() 可以調整子圖之間的間距，使得子圖之間不會重疊。
plt.tight_layout()
plt.show()
```
