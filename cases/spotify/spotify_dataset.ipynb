```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

---
# 檢視資料

```python
# 讀取資料
df = pd.read_csv('./genres_v2.csv', low_memory=False); df
```

```python
# 檢視統計量
df.describe()
```

```python
# 檢查所有欄位的缺失值個數
df.isnull().sum()
```

```python
# 最小最大值標準化 (特徵縮放)
def min_max_scaling(column):
    return (column - column.min()) / (column.max() - column.min())

# 要標準化的欄位
list_columns = [
    'danceability', 'energy', 'key', 'loudness', 'mode',
    'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence',
    'tempo', 'duration_ms', 'time_signature'
]

# 特徵縮放
for col in df[ list_columns ]:
    df[col] = min_max_scaling(df[col])

# 檢視結果
df[ list_columns ]
```

```python
# 丟掉一些不會拿來用的特徵
list_drop = [
    'song_name', 'Unnamed: 0', 'title', 'type', 'mode', 
    'id', 'uri', 'track_href', 'analysis_url'
]

# 將結果存放在新的 dataframe 當中
df_ = df.drop(list_drop, axis=1); df_
```

```python
# 檢視是否還有特徵存在缺失值
df_.isnull().sum()
```

```python
# 選擇可能會用到的特徵，記得包括 label (genre)
list_columns = [
    'danceability', 'energy', 'key', 'loudness',
    'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence',
    'tempo', 'time_signature', 'genre'
]

# 覆蓋原先的 dataframe
df_ = df_[ list_columns ]; df_
```

```python
# 檢視統計量
df_.describe()
```

```python
# 檢視 label 的分布狀況
for genre in df_['genre'].unique().tolist():
    print(f"{genre}: {len(df_[ df_['genre'] == genre ])} 筆")
```

```python
# 將 genre 的值去除重複後，作為 x label
'''
['Dark Trap',
 'Underground Rap',
 'Trap Metal',
 'Emo',
 'Rap',
 'RnB',
 'Pop',
 'Hiphop',
 'techhouse',
 'techno',
 'trance',
 'psytrance',
 'trap',
 'dnb',
 'hardstyle']
'''
x_axis_list = df_['genre'].unique().tolist(); x_axis_list

# 存放 y 值的變數初始化
y_axis_list = []

# 將 genre 對應的 y 值依序放到存放 y 的變數當中
for genre in df_['genre'].unique().tolist():
    y_axis_list.append(len(df_[ df_['genre'] == genre ]))
    

# 設定圖片
plt.title(label="genre in spotify datasets", loc="center", fontsize=14)
plt.xlabel("genre")
plt.xticks(rotation=90)
plt.ylabel("count")

# 長條圖
plt.bar(x_axis_list, y_axis_list)

# 在長條圖上放置文字
'''
參考連結:
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html

matplotlib.pyplot.text(x, y, s, fontdict=None, **kwargs)
'''
for i in range(len(x_axis_list)):
    plt.text(i, y_axis_list[i], y_axis_list[i], ha="center", fontsize=8)

# 顯示圖片
plt.show()
```
